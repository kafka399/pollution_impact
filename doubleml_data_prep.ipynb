```python
import pandas as pd
import numpy as np
```

```python
traffic = pd.read_csv('traffic_full.csv')
```

```python
traffic = traffic[traffic['VEHICULE'] == 'U']
traffic = traffic.drop(['VEHICULE'], axis=1)
```

```python
traffic
```

```python
traffic = traffic.melt(id_vars=['code', 'DATECOM'], value_vars=['P00_01', 'P01_02',
       'P02_03', 'P03_04', 'P04_05', 'P05_06', 'P06_07', 'P07_08', 'P08_09',
       'P09_10', 'P10_11', 'P11_12', 'P12_13', 'P13_14', 'P14_15', 'P15_16',
       'P16_17', 'P17_18', 'P18_19', 'P19_20', 'P20_21', 'P21_22', 'P22_23',
       'P23_24'], var_name='hour', value_name='total')
traffic
```

```python
traffic['hour'] = traffic['hour'].apply(lambda x: x.split('_')[1] + ':00')
```

```python
traffic = traffic.rename(columns={ 'DATECOM': 'date'})
```

```python
traffic['date'] = pd.to_datetime(traffic['date'])
```

```python
traffic
```

```python
air = pd.read_csv('air.denormalized.csv')
```

```python
air['datetime'] = pd.to_datetime(air['date'] + ' ' + air['hour'], format="%d/%m/%Y %H:%M")
air['date'] = pd.to_datetime(air['date'], format="%d/%m/%Y")
air
```

```python
air = air[air['code'].isin(traffic.code.unique())]
```

```python
air = air[air.polLabel=='NO2']
```

```python
air = air.drop_duplicates()
```

```python
air.groupby('hour').size()
```

```python
### we have same day/hour duplicates, averaging
air = air.groupby(['date', 'code', 'hour'])['value'].mean().reset_index()
```

```python
air
```

## Hourly weather data

```python
weather = pd.read_csv('weather_full.csv')
weather['time']= pd.to_datetime(weather['time'])
weather['hour'] = weather['time'].dt.strftime('%H:%M')
weather = weather[['day', 'hour', 'temp', 'wspd', 'prcp', 'coco']]
weather['prcp'][weather.prcp.isna()] = 0
weather
```

### aggregating the classess of the weather

```python
weather['coco'][(weather['coco'] <= 2)] = 1
weather['coco'][(weather['coco'] > 2) & (weather['coco'] < 7)] = 5
weather['coco'][(weather['coco'] >= 7) ] = 7
```

```python
weather = weather.sort_values(by=['day', 'hour']).fillna(method='ffill')
```

```python
#https://dev.meteostat.net/formats.html#weather-condition-codes
weather.groupby('coco').size()
```

```python
weather = weather.rename(columns={ 'day': 'date'})
weather['date']= pd.to_datetime(weather['date'])
weather
```

```python
df = pd.merge(traffic, air, on=['code', 'date', 'hour'])
df = df[df.value>0]
#df[df.total<10]
```

```python
df = pd.merge(df, weather, on=['date', 'hour'])
df
```
